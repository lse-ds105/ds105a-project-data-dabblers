# Wordcloud Visualisation
The purpose of this Jupyter Notebook is to create a wordclouds highlighting the most frequently used words and phrases in the key period of the Gamestop short squeeze.

### Processing Reddit data

```python
import pandas as pd
from wordcloud import STOPWORDS, WordCloud
import string
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

Reading Reddit data from CSV files into a dataframe

```python
df_january = pd.read_csv('../data/reddit_data/Submissions_2021-01_FilteredBySubreddit_GME.csv')
df_february = pd.read_csv('../data/reddit_data/Submissions_2021-02_FilteredBySubreddit_GME.csv')
df_march = pd.read_csv('../data/reddit_data/Submissions_2021-03_FilteredBySubreddit_GME.csv')

df = pd.concat([df_january, df_february, df_march], ignore_index=True)
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
```

### Visualisations

```python
jan_start = '2021-01-01'  
jan_end = '2021-01-31' 
jan_df = df[(df['created_at'] >= jan_start) & (df['created_at'] <= jan_end)]

# Combining all text into one large string
combined_text = ' '.join(jan_df['text'].astype(str))

# Converting to lowercase and removing punctuation
combined_text = combined_text.lower()
combined_text = combined_text.translate(str.maketrans('', '', string.punctuation))

STOPWORDS = set(STOPWORDS)
STOPWORDS.update(['deleted', 'removed', 'nan', 'will', 'think', 'even'])

mask_image = np.array(Image.open('../visualisations/mask_images/controller_silhouette.png'))

wordcloud = WordCloud(
    stopwords=STOPWORDS,
    background_color='white',
    width=800,
    height=600,
    max_words=200,
    colormap='magma', 
    contour_width=1,
    contour_color='steelblue',
    mask=mask_image,  
    prefer_horizontal=0.9, 
    min_font_size=10,
    max_font_size=150,
    random_state=42
).generate(combined_text)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.text(0.51, 0.25, 'January', fontsize=30, color='steelblue', ha='center', va='bottom', fontweight = 'bold', transform=plt.gcf().transFigure)


plt.show()
```
